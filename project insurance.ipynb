#removing null values

```python
col=list(insurance)
```

```python
col
```

```python
for i in col:
    if insurance[i].dtype=='object':
        insurance[i]=insurance[i].fillna(insurance[i].mode()[0])
    else:
         insurance[i]=insurance[i].fillna(insurance[i].mean())
```

```python
insurance.isna().sum()
```
